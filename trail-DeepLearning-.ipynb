```python
import os
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.preprocessing import image
import numpy as np
import xml.etree.ElementTree as ET

# Define the paths to the PASCAL VOC dataset
dataset_path = 'C:/Users/Admin/M22AI608-DeepLearning-Assignment1/VOCdevkit/VOC2007'
jpeg_images_path = os.path.join(dataset_path, 'JPEGImages')
annotations_path = os.path.join(dataset_path, 'Annotations')

# Load pre-trained ResNet50 model
model = tf.keras.applications.ResNet50(weights='imagenet', include_top=False, pooling='avg')

# Function to extract features and class labels from images with corresponding annotations
def extract_features_and_labels(image_paths, annotation_paths, model):
    features = []  # List to store extracted features
    class_labels = []  # List to store corresponding class labels

    for image_path, annotation_path in zip(image_paths, annotation_paths):
        # Load and preprocess the image
        print(image_path)
        img = image.load_img(image_path, target_size=(224, 224))
        img = image.img_to_array(img)
        img = preprocess_input(img)
        img = np.expand_dims(img, axis=0)

        # Extract features using the pre-trained model
        feature = model.predict(img)
        features.append(feature.flatten())

        # Extract class labels from the annotation file
        tree = ET.parse(annotation_path)
        root = tree.getroot()
        for obj in root.findall('object'):
            class_label = obj.find('name').text
            print(class_label," classlabel")
            class_labels.append(class_label)

    features = np.stack(features)
    class_labels = np.array(class_labels)
    return features, class_labels

# Load the image and annotation filenames
image_filenames = sorted(os.listdir(jpeg_images_path))
image_paths = [os.path.join(jpeg_images_path, filename) for filename in image_filenames]
annotation_filenames = sorted(os.listdir(annotations_path))
annotation_paths = [os.path.join(annotations_path, filename) for filename in annotation_filenames]

# Extract features and class labels from the dataset
features, class_labels = extract_features_and_labels(image_paths, annotation_paths, model)

# Print the shape of the extracted features and class labels
print("Features shape:", features.shape)
print("Class labels shape:", class_labels.shape)

```

```python
#Split data into train and valdiation sets
from sklearn.model_selection import train_test_split

# Split the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(features_aligned, labels_binary, test_size=0.2, random_state=42)

# Print the shapes of the training and validation sets
print("Training set shape:", X_train.shape, y_train.shape)
print("Validation set shape:", X_val.shape, y_val.shape)
```

```python
# train binary SVM classifiers for each class
from sklearn.svm import SVC
import joblib

# Train binary SVM classifiers
classifiers = []
for i in range(y_train.shape[1]):
    # Prepare the training data for the current class
    y_train_class = y_train[:, i]

    # Create an SVM classifier
    svm = SVC(kernel='linear')

    # Train the SVM classifier
    svm.fit(X_train, y_train_class)

    # Append the trained classifier to the list
    classifiers.append(svm)

    # Save the trained classifier
    joblib.dump(svm, f'classifier_{i}.joblib')
```

```python
#evaluate the classification accuracy for all the 20 classes on the validation set. Also provide the confusion matrix.

from sklearn.metrics import accuracy_score, confusion_matrix

# Load the saved SVM classifiers
classifiers = []
for i in range(labels_binary.shape[1]):
    svm = joblib.load(f'classifier_{i}.joblib')
    classifiers.append(svm)

# Predict class labels for the validation set
predicted_labels = []
for svm in classifiers:
    predicted_label = svm.predict(X_val)
    predicted_labels.append(predicted_label)

# Convert the predicted labels to a numpy array
predicted_labels = np.array(predicted_labels).T

# Calculate the overall accuracy
overall_accuracy = accuracy_score(y_val, predicted_labels)
print("Overall Accuracy:", overall_accuracy)

# Calculate the accuracy for each class
class_accuracies = []
for i in range(labels_binary.shape[1]):
    class_accuracy = accuracy_score(y_val[:, i], predicted_labels[:, i])
    class_accuracies.append(class_accuracy)

# Print the accuracy for each class
for i, class_accuracy in enumerate(class_accuracies):
    class_name = label_binarizer.classes_[i]
    print(f"Accuracy for class {class_name}: {class_accuracy}")

# Generate the confusion matrix
confusion_mat = confusion_matrix(y_val.ravel(), predicted_labels.ravel())
print("Confusion Matrix:")
print(confusion_mat)
```
